##### kindly find the below code for generating a quanum circuit with a specific number of qubits and gates specified
this is a function named random_circuit which takes 2 integer numbers: 
<br>*number of qubits 
<br>*depth of the circuit
<br>and a list "basis_gates" to return a quantum circuit generated with these specifications
the circuit is randomly generated by looping over the gates propsed in the calling of the function and returns an QuantumCirucit object representing this circuit

```python
from qiskit import QuantumCircuit
import random
```

```python
#list of possible pairs of qubits to interact with
# there can be a better math formula to generate these combinations but I not aware of it 
possible_combinations = [(0,1), (1,2), (2,3), (3,4), (4,2), (5,4),(1,0), (2,1), (3,2), (4,3), (2,4) , (4,5)]

def random_circuit (num_qubits, depth, basis_gates): 
    # intitalizing a quantum ciruit with the number of qubits entered
    circuit = QuantumCircuit(num_qubits)
     # Loop over the specified depth
    for d in range(depth):
        # Loop over each qubit in the circuit
        for q in range(num_qubits):
            # Apply a random single-qubit gate to the qubit
            gate = random.choice(basis_gates)
            while True:       
                q = random.randint(0,num_qubits-1)
                # do while loop to make sure that they are in the possible combination in the range specified by the user
                q1, q2 = random.choice(possible_combinations)
                if q1 < num_qubits and q2 <num_qubits:
                    break
            #decoding the string into the circuits we want
            # any number of gates can be further added and even parameterized circuits
            if gate == 'x':
                circuit.x(q)
            elif gate == 'y':
                circuit.y(q)
            elif gate == 'z':
                circuit.z(q)
            elif gate == 'h':
                circuit.h(q)
            elif gate == 't':
                circuit.t(q)
            elif gate == 'cx':
                    circuit.cx(q1, q2)
            elif gate == 'cz':
                circuit.cz(q1, q2)
            elif gate == 'ch':
                circuit.ch(q1, q2)
            
            # the code to check the depth and layers can be added here

    return circuit
    
    
```

```python
for i in range(10):
    qc = random_circuit(num_qubits=4, depth=3, basis_gates=['x', 'y', 'z', 'cx','cs','ch'])
    print(qc)
    qc.draw()
```

# what is the idea

The function starts by creating a quantum circuit with the specified number of qubits. It then loops over the layers and applying a random single-qubit gate to each qubit in the circuit. At the same time it randomize the one chosen to perform this operation on.
The set of possible single-qubit gates is determined by the input list basis_gates, which can include 'x', 'y', 'z', 'h' and 't'.
then checking for the layer and whether it will continue or not

Finally, the function returns the random quantum circuit.
